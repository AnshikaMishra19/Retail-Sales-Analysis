```python
# ==========================================
# Question 1:
# What are the total sales in each category?
# ==========================================

df.groupby("Category")["Sales"].sum()
```

```python
# ==========================================
# Question 2:
# What is the total profit in each category?
# ==========================================

df.groupby("Category")["Profit"].sum()
```

```python
# ==========================================
# Question 3:
# Which region has the highest sales?
# ==========================================

df.groupby("Region")["Sales"].sum()
```

```python
# ==========================================
# Question 4:
# Which state has the highest sales?
# ==========================================

df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
# ==========================================
# Question 5:
# Which are the Top 10 Customers by Sales?
# ==========================================

df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)
```

```python
# ==========================================
# Question 6:
# Show Sales by Category
# ==========================================

df.groupby("Category")["Sales"].sum().plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
# ==========================================
# Question 7:
# Show Profit by Category
# ==========================================

df.groupby("Category")["Profit"].sum().plot(kind="bar")

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.show()
```

```python
# ==========================================
# Question 8:
# Show Sales by Region
# ==========================================

df.groupby("Region")["Sales"].sum().plot(kind="bar")

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

```python
# ==========================================
# Question 9:
# Show Top 10 States by Sales
# ==========================================

df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10).plot(kind="bar")

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")

plt.show()
```

```python
# ==========================================
# Question 10:
# Show Sales Distribution
# ==========================================

plt.hist(df["Sales"])

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```
